### Getting Players Images

```python
from typing import Optional
from urllib.parse import urlparse, unquote
import requests
from bs4 import BeautifulSoup
```

```python
def wikipedia_page_to_raw_image_url(page_url: str, timeout: int = 20) -> Optional[str]:
    """
    Receives a Wikipedia article URL and returns the original Wikimedia image URL
    from the page infobox, or None if no image is found.

    Example:
    https://en.wikipedia.org/wiki/Rolando_(footballer)
    -> https://upload.wikimedia.org/...
    """
    if not page_url:
        return None

    if "upload.wikimedia.org" in page_url:
        return page_url

    session = requests.Session()
    headers = {
        "User-Agent": "Mozilla/5.0"
    }

    # 1) Load article HTML
    resp = session.get(page_url, headers=headers, timeout=timeout)
    resp.raise_for_status()
    soup = BeautifulSoup(resp.text, "html.parser")

    # 2) Find the image file page link inside the infobox
    file_title = None

    # Most reliable selector on Wikipedia pages
    anchor = soup.select_one("table.infobox a.mw-file-description")
    if anchor and anchor.get("href", "").startswith("/wiki/File:"):
        file_title = unquote(anchor["href"].split("/wiki/", 1)[1])

    # Fallback: find image and climb to parent link
    if not file_title:
        img = soup.select_one("table.infobox img")
        if img:
            parent_a = img.find_parent("a")
            if parent_a and parent_a.get("href", "").startswith("/wiki/File:"):
                file_title = unquote(parent_a["href"].split("/wiki/", 1)[1])

    if not file_title:
        return None

    # 3) Query MediaWiki API to get the original raw image URL
    parsed = urlparse(page_url)
    api_url = f"{parsed.scheme}://{parsed.netloc}/w/api.php"

    params = {
        "action": "query",
        "titles": file_title,
        "prop": "imageinfo",
        "iiprop": "url",
        "format": "json",
    }

    api_resp = session.get(api_url, params=params, headers=headers, timeout=timeout)
    api_resp.raise_for_status()
    data = api_resp.json()

    pages = data.get("query", {}).get("pages", {})
    for page in pages.values():
        imageinfo = page.get("imageinfo")
        if imageinfo:
            return imageinfo[0].get("url")

    return None
```

### Top 5 Scorers

```python
dict_position = {
    "forward": "Atacante",
    "midfielder": "Meio-campista",
    "defender": "Defensor",
    "goal_keeper": "Goleiro",
}

dict_golden_ball = {
    "Golden Ball": "Bola de Ouro",
    "Silver Ball": "Bola de Prata",
    "Bronze Ball": "Bola de Bronze",
}
```

```python
def make_player_name(df):
    return (
        df["given_name"].fillna("").str.strip()
        .str.cat(df["family_name"].fillna("").str.strip(), sep=" ")
        .str.replace(r"\s+", " ", regex=True)
        .str.replace("not applicable", "", regex=False)
        .str.strip()
    )

g = goals.copy()
g["player_name"] = make_player_name(g)
```

```python
# Create position column by extracting from goal_keeper, defender, midfielder, forward columns
def get_position(row):
    if row['goal_keeper']:
        return 'goal_keeper'
    elif row['defender']:
        return 'defender'
    elif row['midfielder']:
        return 'midfielder'
    elif row['forward']:
        return 'forward'
    else:
        return 'unknown'

if "position" not in players.columns:
    players['position'] = players.apply(get_position, axis=1)

players["posição"] = players["position"].map(dict_position)

if "goal_keeper" in players.columns: 
    players = players.drop(columns=['goal_keeper', 'defender', 'midfielder', 'forward','female'])
    
players.head(10)
```

```python
# Usually own goals should not count for the scorer. If own_goal is 1, we should not count the goal for the player.
#  If own_goal is 0, we can assume it's not an own goal.
g = g[g["own_goal"] != 1].copy()
      
goal_totals = (
    g.groupby(
        ["year","tournament_name", "player_name", "player_team_name", "player_team_code", "player_id"],
        as_index=False
    )
    .agg(goals=("key_id", "size"))
)

goal_totals = goal_totals.sort_values(
    ["tournament_name", "goals", "player_name"],
    ascending=[True, False, True]
)

goal_totals["rank"] = (
    goal_totals.groupby("tournament_name")["goals"]
    .rank(method="first", ascending=False).astype(int)
)

top5 = goal_totals[goal_totals["rank"] <= 5].copy()

top5 = top5.merge(
    players[["player_id", "player_wikipedia_link", "posição"]],
    on="player_id",
    how="left"
)

#top5["wikipedia_profile_link"] = top5["player_wikipedia_link"].apply(wikipedia_page_to_raw_image_url)
```

### Golden Ball / Silver Ball / Bronze Ball Players


```python
podium_awards = awards[awards["award_name"].isin(["Golden Ball", "Silver Ball", "Bronze Ball"])].copy()
podium_awards["player_name"] = make_player_name(podium_awards)
podium_awards.head(5)
```

```python
golden_ball = podium_awards.merge(
    players[["player_id", "player_wikipedia_link", "posição"]],
    on="player_id",
    how="left"
)

golden_ball["premio"] = golden_ball["award_name"].map(dict_golden_ball)

if "given_name" in golden_ball.columns:
    golden_ball = golden_ball.drop(columns=['given_name', 'family_name', 'shared', 'award_name', 'tournament_id'])

#golden_ball["wikipedia_profile_link"] = golden_ball["player_wikipedia_link"].apply(wikipedia_page_to_raw_image_url)
golden_ball.head(5)
```
