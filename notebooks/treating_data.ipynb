```python
Gender = "Men"
if Gender == "Men":
    matches = matches[matches["tournament_name"].str.contains("Men's World Cup")]
    goals = goals[goals["tournament_name"].str.contains("Men's World Cup")]
else:
    matches = matches[matches["tournament_name"].str.contains("Women's World Cup")]
    goals = goals[goals["tournament_name"].str.contains("Women's World Cup")]
```
